## Чтение и отображение изображений

Для чтения изображения с помощью OpenCV мы можем использовать функцию `cv2.imread()`. Эта функция принимает в качестве аргумента путь к изображению и возвращает массив NumPy, содержащий данные изображения. По умолчанию изображение считывается в цветном режиме.

Мы выводим изображение с помощью функции `cv2_imshow()` – патча для Google Colab. Встроенная функция OpenCV `cv2.imshow()` часто вызывает зависания Jupyter Notebook.

```python
# считаем изображение
img = cv2.imread('image.jpg')

# выведем изображение
plt.imshow(img);
```

Видим, что картинка отличается от оригинальной! 

OpenCV интерпретирует данные изображения, используя BGR (Blue, Green, Red) вместо RGB (Red, Green, Blue), поскольку исторически BGR был порядком цветов, по умолчанию используемым в растровых файлах Windows, которые были одним из самых ранних форматов файлов, поддерживаемых OpenCV.

**Пиксель** – это наименьшая единица цифрового изображения. Каждый пиксель обычно представлен комбинацией числовых значений, которые представляют интенсивность или цвет пикселя. Количество каналов в изображении представляет собой число компонентов, используемых для представления каждого пикселя.

В полутоновом изображении каждый пиксель представлен одним числовым значением, которое представляет интенсивность пикселя. Поэтому изображение в градациях серого имеет только один канал. *Изображение в оттенках серого хранит только одно значение - целое число от 0 до 255.*

В цветном изображении каждый пиксель представлен тремя числовыми значениями, которые представляют интенсивность красного, зеленого и синего компонентов пикселя. Эти три компонента обычно называются каналами RGB. *Цветное изображение хранит три значения: три целых числа от 0 до 255.*

Некоторые форматы изображений, такие как RGBA (красный, зеленый, синий и альфа) или CMYK (голубой, пурпурный, желтый и черный), используют более трех каналов для представления каждого пикселя.

## Изменение размера

Чтобы изменить размер изображения, мы можем использовать функцию `resize()`. Функция принимает два аргумента: изображение, размер которого нужно изменить, и новый размер изображения.

```python
image.shape[0]
```

```python
# уменьшим изображение вполовину – то есть зададим высоту и ширину как 50% от исходного изображения

height = int(image.shape[0] * 0.5) # атрибут shape переменной img возвращает кортеж (height, width, channels), 
                                 # где img.shape[0] - высота изображения
width = int(image.shape[1] * 0.5) # аналогично img.shape[1] дает ширину изображения

resized_img = cv2.resize(image, (width, height)) # функция cv2.resize() принимает два аргумента: 
                                                # изображение и новые размеры в формате кортежа (ширина, высота). 
```

```python
# plt.figure(figsize = (20, 3));
plt.imshow(resized_img); # несмотря на то, что matplotlib отрисовывает картинку в таком же размере, 
                         # на самом деле она уменьшилась (посмотрите на оси!)
```

## Вращение изображений

Чтобы повернуть изображение, мы можем использовать функцию `rotate()`. Функция принимает два аргумента: изображение, которое нужно повернуть, и `rotateCode` -- флаги для ротации (`cv2.ROTATE_90_CLOCKWISE`, `cv2.ROTATE_180`, `cv2.ROTATE_90_COUNTERCLOCKWISE`).

```python
rotated_img = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated_img);
```

Для более сложных вращений придется делать более сложные вычисления (или использовать библиотеку `scipy`).

```python
(h, w, c) = image.shape

M = cv2.getRotationMatrix2D((w/2, h/2), -45, 0.5)
# Cтрока выше создает двумерную матрицу вращения M с помощью функции cv2.getRotationMatrix2D() 
# Функция принимает три аргумента: центральную точку вращения (заданную как (w/2, h/2) для вращения вокруг центра изображения), 
# угол поворота (заданный как 45 градусов в этом примере. Положительное число – поворот против часовой стрелки, отрицательное – по часовой) 
# и коэффициент масштабирования (заданный как 0.5 для сохранения изображения без обрезания).

rotated_img = cv2.warpAffine(image, M, (w, h))
# Эта строка применяет матрицу вращения M к входному изображению img с помощью функции cv2.warpAffine()
# Функция принимает три аргумента: входное изображение, матрицу преобразования M 
# и размер выходного изображения.

plt.imshow(rotated_img);
```

## Размытие изображения

Чтобы размыть изображение, можно использовать функции `cv2.GaussianBlur()` или `cv2.blur()`.

`cv2.GaussianBlur()` принимает следующие аргументы:
1. изображение,
2. кортеж из 2 положительных нечётных чисел (размер шаблона фильтра Гаусса). Чем больше числа, тем больше сила сглаживания.
3. стандартное отклонение гауссовского ядра (по умолчанию 0). Более высокое значение стандартного отклонения приводит к большему размытию, а более низкое значение стандартного отклонения - к меньшему размытию.

```python
blurred_img = cv2.GaussianBlur(resized_img, (11, 11), 0)

plt.imshow(blurred_img);
```

`cv2.blur()` принимает два аргумента: входное изображение и кортеж из двух чисел. 

Функция использует простую технику усреднения для размытия изображения.

```python
blurred_img = cv2.blur(resized_img, (11, 11))

plt.imshow(blurred_img);
```

## А теперь положим картинку на "зеленку"

Или приведем мультяшного питона в Вышку.

Для начала загрузим картинки питона и Атриума Вышки на Покровском бульваре.

```python
python = cv2.cvtColor(cv2.imread("cartoon-python.jpg"), cv2.COLOR_BGR2RGB)

plt.imshow(python);
```

```python
hse = cv2.cvtColor(cv2.imread("hse.jpg"), cv2.COLOR_BGR2RGB)

plt.imshow(hse);
```

Посмотрим на размеры. Видим, что картинка с питоном больше. Поменяем ее размер, чтобы она вписалась в нашу картинку.

```python
print(hse.shape)
print(python.shape)
```

```python
small_python = cv2.resize(python, (2500, 1663))

plt.imshow(small_python);
```

```python
print(hse.shape)
print(small_python.shape)
```

Дальше мы накладываем маску, чтобы убрать зеленый цвет и забрать с картинки только питона.

```python
hsv = cv2.cvtColor(small_python, cv2.COLOR_BGR2HSV) # переводим картинку в HSV (Hue – тон, Saturation – насыщенность и Value – значение.)
mask = cv2.inRange(hsv, (36, 25, 25), (70, 255, 255)) # создаем маску зеленого цвета в формате HSV https://stackoverflow.com/questions/47483951/how-can-i-define-a-threshold-value-to-detect-only-green-colour-objects-in-an-ima

plt.imshow(mask, cmap='gray');
```

```python
# маскируем картинку, чтобы было видно только питона
masked_image = np.copy(small_python) # делаем копию, чтобы оригинал всегда был под рукой -- вдруг где-то накосячим!
masked_image[mask != 0] = [0, 0, 0] # "блокируем" все, кроме питона, красим в черный [0, 0, 0]
                                    # дальше будем складывать массивы, соответственно [0, 0, 0] + [10, 25, 150] = [10, 25, 150]

plt.imshow(masked_image);
```

```python
masked_background_image = np.copy(hse)

# маскируем картинку, но оставляем место под питона (красим в черный)
masked_background_image[mask == 0] = [0, 0, 0]

plt.imshow(masked_background_image);
```

```python
complete_image = masked_image + masked_background_image # складываем массивы

plt.imshow(complete_image);
```

```python
cv2.imwrite('hse_python.png', cv2.cvtColor(complete_image, cv2.COLOR_RGB2BGR))
```
