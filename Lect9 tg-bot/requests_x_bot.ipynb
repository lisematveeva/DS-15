```python
import requests
from bs4 import BeautifulSoup
import json
from tqdm import tqdm
import numpy as np
import pandas as pd
from time import sleep
from collections import Counter
```

```python
vacancies = []

v = requests.get(f"https://api.hh.ru/vacancies?text=data%20аналитик&only_with_salary=true&per_page=100")
num = json.loads(v.text)['found']
num = round(num/100)
#print(num)

for i in range(0, num+1):
    vac = requests.get(f"https://api.hh.ru/vacancies?text=data%20аналитик&only_with_salary=true&per_page=100&page={i}")
    print(i)
    vacancies.extend(json.loads(vac.text)['items'])

print(len(vacancies))
```

```python
cnt = 0
info2 = []
for i in vacancies:
    try:
        name = i['name']
    except:
        name = np.nan
    try:
        money_from = i['salary'].get('from', np.nan)
    except:
        money_from = np.nan
    try:
        money_to = i['salary'].get('to', np.nan)
    except:
        money_to = np.nan
        
    try:
        curr = i['salary'].get('currency', np.nan)
    except:
        curr = np.nan
    
    try:
        area = i['area'].get('name', np.nan)
    except:
        area = np.nan
        
    try:
        emp = i['employer'].get('name', np.nan)
    except:
        emp = np.nan
    url = i['alternate_url']


    info2.append([name, money_from, money_to, curr, area, emp, url])
    
    cnt += 1
    print(cnt)
```

```python
info2
```

```python
df_v = pd.DataFrame(info2, columns = ['name', 'money_from', 'money_to', 'curr', 'area', 'emp', 'url'])
df_v.sort_values('money_from', ascending = False)
```
